## Notes on implemented functions

### Converting timeseries into angular scale

We denote a vector of N directional observations $\alpha_i$ as $\alpha = (\alpha_1,..., \alpha_N)$. To convert angles in degrees to angles in radians:

$$\alpha_{degree}=360deg.\frac{\alpha_{radians}}{2\pi}$$

Data such as **time** of the day can be converted to a common **angular scale in radians** by:

$$\alpha=\frac{2\pi x}{k}$$

where $x$ is the representation of the data in the original scale, $\alpha$ is its angular direction and $k$ is the total number of steps on the scale that y is measured in. For example, we have $x$ representing hour of the day and thus $k = 24$.

### Circular statistics
- Code inspired by Matlab circular statistics toolbox by Dr. Philipp Berens (file:///tmp/mozilla_domi0/v31i10.pdf)
- **Resultant vector:** Reflect the phase consistency of a population of angles. 
It is constructed by vectorial averaging of phase vectors each of unitary length equal to one. 
Thus, the angle of the mean resultant vector represents the mean angle of the population and its length (Phase-Locking Value, PLV) represents dispersion (values close to zero)
or concentration (values close to one) of constituting angles. 
- Given that the PLV is a continuous variable linearly varying from zero to one, we used a **Wilcoxon test** to compare
population medians.

 



**Phase analyses** were done using the 2012 Matlab circular
statistics toolbox by Dr. Philipp Berens including functions for **circular mean,
circular standard deviation, and resultant vector**.

We used the **circular variant of kmeans
for clustering of angles**. We used the **Omnibus (or Hodges-Ajne) test to
calculate statistical significance for non-uniform angular distribution (against the
null hypothesis of a uniform distribution)**, as opposed to the more classical Rayleigh
test, because some angular distributions were bimodal, especially in circadian
rhythms. We used the **Kuiper test (analog to the Kolmogorovâ€“Smirnov** test for
circular data) to calculate the statistical difference between two angle populations,




### Mean ... do it in exp(i,j)
directions are first transformed to unit vectors in the two-dimensional plane by:
$$r_i=\begin{pmatrix} \cos \alpha_i \\ \sin \alpha_i \end{pmatrix}$$

**mean resultant vector**: $$\bar{r}=\frac{1}{N}\sum_i(r_i)$$

**mean angle**: $$\bar{\alpha}=\displaystyle \operatorname {atan2} (\frac{1}{N}\sum_{j=1}^N{r\sin\alpha_j},\frac{1}{N}\sum_{j=1}^N{r\cos\alpha_j}$$

                     
**length - magnitude**: $$R=||\bar{r}||$$

 $$||r||=\sqrt{(r*\cos\alpha)^2+(r*\sin\alpha)^2}$$
 
 
 
 **Autocorrelation**, also known as serial correlation, is the correlation of a signal with a delayed copy of itself as a function of delay. Informally, it is the similarity between observations as a function of the time lag between them. The analysis of autocorrelation is a mathematical tool for finding repeating patterns, such as the presence of a periodic signal obscured by noise, or identifying the missing fundamental frequency in a signal implied by its harmonic frequencies. It is often used in signal processing for analyzing functions or series of values, such as time domain signals.

## Time-frequency analysis using Morlet Wavelet Transform (pywavelets)

Spectogram:
- bright regions correspond to frequencies of higher power (for a specific amount of time)
- Cone of influence in gray

Periodogram:
- average power per period in signal (purple) and surrogate data (gray) 
- peaks above surrogate line are of interest
- circadian rhythm present (period = 1)
- peaks at ~5 and ~10 days. 10 day peak is possible an artifact due to increased amplitude in raw data of the signal during this period. 

Raw data + filtered data:
- to illustrate surrogate data (shuffled data points)

## Epileptic rat : Inlfuence of absence of light on circadian cycle of Interictal Spikes, Temperature and Activity

- x axis: Hour of the day
- y axis: Resultant vector 
- points are set to 1 and jittered

'normal'
- During 7am/7pm light cycle low and high values (of Spikes, Temperature & Activity) consistently occur during animal's sleep and active hours, respectively.  
- Resultant vector > 0.6 

'dark_only'
- During constant dark, rat's circadian rhythm is driven by the internal hypothalamic clock. 
- the values are dispersed => deviation from 24h cycle
